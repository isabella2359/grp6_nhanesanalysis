#Background
Our health issue touches on how thyroid hormone, which regulate metabolism, affects sleep. The two main hormones released by the thyroid are thyroxine (T4) and triiodothyronine (T3) (Jonklaas et al., 2014). The thyroid stimulating hormone (TSH) made by the pituitary gland in our brain stimulates the thyroid to produce T3 and T4 hormones, released into our bloodstream.

Free T4 ‚Äì The word ‚ÄúFree" in its name describes that this hormone is not bound to proteins in the blood and is actively flowing in the body. It functions in controlling metabolism, energy use, and body temperature (Jonklaas et al., 2014).

Total T3 ‚ÄìThe active form of thyroid hormone that affects almost every physiological process. The word ‚ÄúTotal‚Äù in its name means that it includes both the protein-bound and free (active) forms of T3. Its functions include boosting metabolism, heart rate, and body temperature, with stronger effects than that of T4 (Jonklaas et al., 2014).

Hypothyroidism (underactive thyroid, usually associated with low levels of T3 and T4) is a common condition of thyroid hormone deficiency, which is readily diagnosed and managed but potentially fatal in severe cases, if left untreated (Chaker et al ,2017).

Hypothyroidism can present with a wide range of symptoms, from severe and life threatening to no noticeable signs. The typical common symptoms in adults are fatigue, lethargy, cold intolerance, weight gain, constipation, change in voice, and dry skin. However, symptoms can differ with age and sex, among other factors. While the standard treatment for patients is thyroid hormone replacement therapy with levothyroxine, they may experience symptoms even after their hormone levels return to normal (Chaker et al ,2017).

People with hypothyroidism, often experience trouble tolerating cold at night and joint or muscle pain that disrupts sleep. Research links hypothyroidism to‚ÄØpoor quality sleep, longer time to fall asleep, and shorter sleep duration. Younger individuals, women, and those with a relatively low body mass index, are at higher risk of developing sleep problems due to hypothyroidism (Sleep Foundation, 2024).

Conversely, hyperthyroidism (overactive thyroid, associated with high levels of T4 and T3) can also cause sleep difficulties through nervousness, irritability, muscle weakness and constant feelings of tiredness. Additionally, individuals with overactive thyroid may experience night sweats and frequent urges to urinate, both of which can disrupt sleep (Sleep Foundation, 2024).

Overall, the association between sleep and hormones has long been well established. Research suggests that deep sleep significantly influences the functioning of the endocrine system. For example, the secretions of growth hormone (GH) and prolactin (PRL) increase during deep sleep, while cortisol and thyrotropin (TSH) secretions decrease. Some studies have shown how the inflammatory system can be affected by changes in the circadian rhythm caused by sleep disruption and integrity (Nazem et al., 2021). (Nazem et al., 2021)

While many researchers have indicated that hypothyroidism and hyperthyroidism can result in sleep disturbances such as hypersomnia. Our research expands on this by incorporating three key thyroid hormones (TSH, Free T4, and Total T3) and exploring their associations not only with sleep duration, but also with depression scores (PHQ-9) and BMI, both of which are independently and jointly relevant to thyroid regulation and general well-being. By taking a multidimensional approach, our study seeks to provide a more comprehensive understanding of how thyroid function relates to physical and mental health in the general U.S. adult population, as a perspective not fully addressed in prior research. ## Creating the Data Frame


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# NHANES years and cycles (2007‚Äì2012)
years = ["2007", "2009", "2011"]
cycles = ["E", "F", "G"]

# Helper function to load NHANES datasets
def load_nhanes_data(year, cycle, prefix, required_cols):
    url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{prefix}_{cycle}.xpt"
    try:
        df = pd.read_sas(url)
        cols_present = [c for c in required_cols if c in df.columns]
        return df[cols_present]
    except Exception as e:
        print(f"‚ùå Failed to load {prefix}_{cycle}: {e}")
        return pd.DataFrame(columns=required_cols)

# Define variables per dataset
vars_demo = ["SEQN", "RIAGENDR", "RIDAGEYR"]
vars_bmx = ["SEQN", "BMXBMI"]
# Include LBXTSH1 as fallback
vars_thyroid = ["SEQN", "LBXTSH", "LBXTSH1", "LBXTT3", "LBXT4F"]
vars_slq = ["SEQN", "SLD010H", "SLQ060"]
vars_paq = ["SEQN", "PAD615"]
vars_smq = ["SEQN", "SMQ020"]
vars_depression = ["SEQN", "DPQ010", "DPQ020", "DPQ030", "DPQ040", "DPQ050", "DPQ060", "DPQ070", "DPQ080", "DPQ090"]

df_list = []
for y, c in zip(years, cycles):
    df_demo = load_nhanes_data(y, c, "DEMO", vars_demo)
    df_bmx = load_nhanes_data(y, c, "BMX", vars_bmx)
    df_thyroid = load_nhanes_data(y, c, "THYROD", vars_thyroid)

    # Fix for TSH column naming
    if "LBXTSH" not in df_thyroid.columns and "LBXTSH1" in df_thyroid.columns:
        df_thyroid.rename(columns={"LBXTSH1": "LBXTSH"}, inplace=True)

    df_slq = load_nhanes_data(y, c, "SLQ", vars_slq)
    df_paq = load_nhanes_data(y, c, "PAQ", vars_paq)
    df_smq = load_nhanes_data(y, c, "SMQ", vars_smq)
    df_dpq = load_nhanes_data(y, c, "DPQ", vars_depression)

    dfs = [df_demo, df_bmx, df_thyroid, df_slq, df_paq, df_smq, df_dpq]
    df_merged = dfs[0]
    for d in dfs[1:]:
        df_merged = pd.merge(df_merged, d, on="SEQN", how="outer")

    df_merged["cycle"] = c
    df_merged["year"] = y
    df_list.append(df_merged)

# Combine all years
df_final = pd.concat(df_list, ignore_index=True)

# Calculate PHQ-9 depression score (sum of 9 DPQ items)
phq_items = ["DPQ010", "DPQ020", "DPQ030", "DPQ040", "DPQ050", "DPQ060", "DPQ070", "DPQ080", "DPQ090"]
df_final["Depression"] = df_final[phq_items].sum(axis=1, skipna=False)

# Rename columns for clarity
df_final.rename(columns={
    "RIAGENDR": "Gender",
    "RIDAGEYR": "Age",
    "BMXBMI": "BMI",
    "LBXTSH": "TSH",
    "LBXTT3": "TotalT3",
    "LBXT4F": "FreeT4",
    "SLD010H": "SleepDuration",
    "SLQ060": "SleepQuality",
    "PAD615": "PhysicalActivity",
    "SMQ020": "SmokingStatus"
}, inplace=True)

# Create your final analysis dataframe with selected columns
df_analysis = df_final[[
    "SEQN",
    "Gender",
    "Age",
    "BMI",
    "TSH",
    "TotalT3",
    "FreeT4",
    "SleepDuration",
    "SleepQuality",
    "PhysicalActivity",
    "SmokingStatus",
    "Depression",
    "cycle",
    "year"
]].copy()

# Preview the dataframe
print(df_analysis.head())
```

## Data Cleaning and Handling of Missing Data

```python
# ------------------------
# NHANES Data Cleaning Script
# ------------------------

import numpy as np

# PHQ-9 items (Depression scale)
phq_items = ["DPQ010", "DPQ020", "DPQ030", "DPQ040", "DPQ050", "DPQ060", "DPQ070", "DPQ080", "DPQ090"]

# Categorical columns
cat_cols = ["Gender", "SmokingStatus", "SleepQuality"]
for col in cat_cols:
    df_final[col] = df_final[col].astype("category")

# Drop implausible SleepDuration and BMI values
df_final = df_final[(df_final["SleepDuration"] >= 0) & (df_final["SleepDuration"] <= 24)]
df_final = df_final[(df_final["BMI"] >= 10) & (df_final["BMI"] <= 80)]

# Recode invalid PHQ-9 values (9 ‚Üí NaN)
for item in phq_items:
    df_final[item] = df_final[item].replace(9, np.nan)

# Recode invalid PhysicalActivity values
invalid_pa_codes = [9999]
df_final["PhysicalActivity"] = df_final["PhysicalActivity"].replace(invalid_pa_codes, np.nan)

# Compute depression score (only if all items are present)
df_final["Depression"] = df_final[phq_items].sum(axis=1, skipna=False)

# Show missing data count before dropping
print("\nüìå Missing Data Before Cleaning:")
print(df_final[["SleepDuration", "BMI", "Depression", "TSH", "TotalT3", "FreeT4", "Age",
                "Gender", "SmokingStatus", "PhysicalActivity"]].isnull().sum().sort_values(ascending=False))

# Define key variables needed for modeling/statistics
key_vars = ["SleepDuration", "BMI", "Depression", "TSH", "TotalT3", "FreeT4",
            "Age", "Gender", "SmokingStatus", "PhysicalActivity"]

# Drop rows with missing data in any key variable
df_clean = df_final.dropna(subset=key_vars).reset_index(drop=True)

# Cap extreme hormone values (outliers)
df_clean.loc[df_clean["TSH"] > 20, "TSH"] = 20
df_clean.loc[df_clean["TotalT3"] > 300, "TotalT3"] = 300
df_clean.loc[df_clean["FreeT4"] > 4, "FreeT4"] = 4

# Show dataset shape and confirm all missing values removed
print(f"\n‚úÖ Final cleaned dataset shape: {df_clean.shape}")
print("\n‚úÖ Missing values after cleaning (should all be 0):")
print(df_clean[key_vars].isnull().sum())
```

##Shapiro-Wilk Normality Test

```python
from scipy.stats import shapiro
import pandas as pd

# Define numeric columns you want to test
numeric_cols = ["Age", "BMI", "TSH", "TotalT3", "FreeT4", "SleepDuration", "PhysicalActivity"]

# Create empty list to collect results
shapiro_results = []

# Run test for each variable
for col in numeric_cols:
    data = df_clean[col].dropna()
    if len(data) > 3:  # Shapiro requires at least 3 non-NaN values
        sample_data = data.sample(n=min(len(data), 5000), random_state=42)
        stat, p = shapiro(sample_data)
        interpretation = "Normal" if p > 0.05 else "Non-normal"
        shapiro_results.append({
            "Variable": col,
            "W Statistic": stat,
            "p-value": p,
            "Normality": interpretation
        })
    else:
        shapiro_results.append({
            "Variable": col,
            "W Statistic": None,
            "p-value": None,
            "Normality": "Insufficient data"
        })

# Create dataframe to display
norm_df = pd.DataFrame(shapiro_results)

# Format p-values for readability
norm_df["p-value"] = norm_df["p-value"].apply(lambda x: f"{x:.4f}" if pd.notnull(x) else "NA")

# Show the result
print("Shapiro-Wilk Normality Test Results:")
display(norm_df)


```

All variables failed the normality test (p < 0.05).

Most of them are strongly non-normal, especially TSH and FreeT4 (very low W values, consistent with extreme skewness)and Physical Activity (the most non-normal). Even Age and Sleep Duration, which looked mildly skewed, are statistically non-normal due to large sample size.

```python
import pandas as pd
import matplotlib.pyplot as plt

# STEP 1: Create DepressionScore from PHQ-9 items
phq_items = ['DPQ010', 'DPQ020', 'DPQ030', 'DPQ040', 'DPQ050', 'DPQ060', 'DPQ070', 'DPQ080', 'DPQ090']

# Replace invalid PHQ-9 responses (7=Refused, 9=Don't know) with NaN
df_clean[phq_items] = df_clean[phq_items].replace([7, 9], pd.NA)

# Calculate total PHQ-9 score
df_clean['Depression'] = df_clean[phq_items].astype('float').sum(axis=1)

# Categorize depression severity
df_clean['DepressionScore'] = pd.cut(
    df_clean['Depression'],
    bins=[-1, 4, 9, 14, 27],  # PHQ-9 cutoffs
    labels=[0, 1, 2, 3]       # 0=None/Minimal, 1=Mild, 2=Moderate, 3=Severe
).astype('float')

# STEP 2: Define categorical variables and label mappings
cat_cols = ['Gender', 'SmokingStatus', 'DepressionScore']

category_labels = {
    'Gender': {1.0: 'Male', 2.0: 'Female'},
    'SmokingStatus': {
        1.0: 'Smoker',
        2.0: 'Non-smoker',
        7.0: 'Refused',
        9.0: 'Don‚Äôt know'
    },
    'DepressionScore': {
        0: 'None/Minimal',
        1: 'Mild',
        2: 'Moderate',
        3: 'Severe'
    }
}

# STEP 3: Plot bar charts
for var in cat_cols:
    if var in df_clean.columns:
        raw_data = df_clean[var].astype("float")

        # Map numerical codes to category names
        mapped_data = raw_data.map(category_labels.get(var, {}))

        # Drop unknowns for SmokingStatus
        if var == 'SmokingStatus':
            mapped_data_clean = mapped_data[~mapped_data.isin(['Refused', 'Don‚Äôt know'])].dropna()
        else:
            mapped_data_clean = mapped_data.dropna()

        # Enforce category order for DepressionScore
        if var == 'DepressionScore':
            ordered_categories = ['None/Minimal', 'Mild', 'Moderate', 'Severe']
            mapped_data_clean = pd.Categorical(mapped_data_clean, categories=ordered_categories, ordered=True)

        # Calculate proportions
        proportions = pd.Series(mapped_data_clean).value_counts(normalize=True).sort_index()

        # Plot
        plt.figure(figsize=(6, 4))
        bars = proportions.plot(kind='bar', color='lightgreen', edgecolor='black')

        # Add percentage labels
        for i, val in enumerate(proportions):
            plt.text(i, val + 0.01, f"{val:.2%}", ha='center', va='bottom')

        plt.title(f"Proportion of {var}")
        plt.ylabel("Proportion")
        plt.xticks(rotation=0)
        plt.ylim(0, proportions.max() + 0.1)
        plt.tight_layout()
        plt.show()
    else:
        print(f"Warning: {var} not found in dataframe columns.")





```

## Linear Regression

```python
#linear regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

biomarkers = ['SleepDuration', 'BMI', 'Depression']
cont_predictors = ['TSH', 'TotalT3', 'FreeT4', 'Age', 'PhysicalActivity']

for biomarker in biomarkers:
    if biomarker not in df_final.columns:
        print(f"Warning: {biomarker} not found in dataframe columns. Skipping.")
        continue

    fig, axs = plt.subplots(1, len(cont_predictors), figsize=(5 * len(cont_predictors), 4), squeeze=False)

    for i, predictor in enumerate(cont_predictors):
        if predictor not in df_final.columns:
            print(f"Warning: {predictor} not found in dataframe columns. Skipping.")
            continue

        data = df_final[[biomarker, predictor]].dropna()
        if data.empty:
            print(f"No data for {biomarker} vs {predictor}, skipping plot.")
            continue

        X = data[[predictor]].values
        y = data[biomarker].values

        # Fit linear regression
        model = LinearRegression().fit(X, y)
        y_pred = model.predict(X)
        r2 = r2_score(y, y_pred)

        # Sort for plotting smooth line
        sorted_idx = X.flatten().argsort()
        X_sorted = X[sorted_idx]
        y_pred_sorted = y_pred[sorted_idx]

        ax = axs[0, i]
        ax.scatter(X, y, alpha=0.5)
        ax.plot(X_sorted, y_pred_sorted, color='red')
        ax.set_title(f'{biomarker} vs {predictor}\n$R^2$ = {r2:.3f}')
        ax.set_xlabel(predictor)
        ax.set_ylabel(biomarker)
        ax.grid(True)

    plt.tight_layout()
    plt.show()
```

Our key takeaway from here is that there are no strong predictors as none of the independent variables (TSH, TotalT3, FreeT4, Age, PhysicalActivity) show meaningful linear relationships with BMI, SleepDuration, or Depression (all R¬≤ values are near zero).

Our potential next steps include investigating possible non-linear relationships (e.g., logarithmic or polynomial models), exploring additional variables not currently included, and checking data quality, including outliers and measurement errors.
