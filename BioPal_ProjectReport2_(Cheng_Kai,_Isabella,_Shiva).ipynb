#Background
Our health issue touches on how thyroid hormone, which regulate metabolism, affects sleep. The two main hormones released by the thyroid are thyroxine (T4) and triiodothyronine (T3) (Jonklaas et al., 2014). The thyroid stimulating hormone (TSH) made by the pituitary gland in our brain stimulates the thyroid to produce T3 and T4 hormones, released into our bloodstream.

Free T4 ‚Äì The word ‚ÄúFree" in its name describes that this hormone is not bound to proteins in the blood and is actively flowing in the body. It functions in controlling metabolism, energy use, and body temperature (Jonklaas et al., 2014).

Total T3 ‚ÄìThe active form of thyroid hormone that affects almost every physiological process. The word ‚ÄúTotal‚Äù in its name means that it includes both the protein-bound and free (active) forms of T3. Its functions include boosting metabolism, heart rate, and body temperature, with stronger effects than that of T4 (Jonklaas et al., 2014).

Hypothyroidism (underactive thyroid, usually associated with low levels of T3 and T4) is a common condition of thyroid hormone deficiency, which is readily diagnosed and managed but potentially fatal in severe cases, if left untreated (Chaker et al ,2017).

Hypothyroidism can present with a wide range of symptoms, from severe and life threatening to no noticeable signs. The typical common symptoms in adults are fatigue, lethargy, cold intolerance, weight gain, constipation, change in voice, and dry skin. However, symptoms can differ with age and sex, among other factors. While the standard treatment for patients is thyroid hormone replacement therapy with levothyroxine, they may experience symptoms even after their hormone levels return to normal (Chaker et al ,2017).

People with hypothyroidism, often experience trouble tolerating cold at night and joint or muscle pain that disrupts sleep. Research links hypothyroidism to‚ÄØpoor quality sleep, longer time to fall asleep, and shorter sleep duration. Younger individuals, women, and those with a relatively low body mass index, are at higher risk of developing sleep problems due to hypothyroidism (Sleep Foundation, 2024).

Conversely, hyperthyroidism (overactive thyroid, associated with high levels of T4 and T3) can also cause sleep difficulties through nervousness, irritability, muscle weakness and constant feelings of tiredness. Additionally, individuals with overactive thyroid may experience night sweats and frequent urges to urinate, both of which can disrupt sleep (Sleep Foundation, 2024).

Overall, the association between sleep and hormones has long been well established. Research suggests that deep sleep significantly influences the functioning of the endocrine system. For example, the secretions of growth hormone (GH) and prolactin (PRL) increase during deep sleep, while cortisol and thyrotropin (TSH) secretions decrease. Some studies have shown how the inflammatory system can be affected by changes in the circadian rhythm caused by sleep disruption and integrity (Nazem et al., 2021). (Nazem et al., 2021)

# Section 6: Variables and Data Sources

## Creating the Data Frame


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# NHANES years and cycles (2007‚Äì2012)
years = ["2007", "2009", "2011"]
cycles = ["E", "F", "G"]

# Helper function to load NHANES datasets
def load_nhanes_data(year, cycle, prefix, required_cols):
    url = f"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/{year}/DataFiles/{prefix}_{cycle}.xpt"
    try:
        df = pd.read_sas(url)
        cols_present = [c for c in required_cols if c in df.columns]
        return df[cols_present]
    except Exception as e:
        print(f"‚ùå Failed to load {prefix}_{cycle}: {e}")
        return pd.DataFrame(columns=required_cols)

# Define variables per dataset
vars_demo = ["SEQN", "RIAGENDR", "RIDAGEYR"]
vars_bmx = ["SEQN", "BMXBMI"]
# Include LBXTSH1 as fallback
vars_thyroid = ["SEQN", "LBXTSH", "LBXTSH1", "LBXTT3", "LBXT4F"]
vars_slq = ["SEQN", "SLD010H", "SLQ060"]
vars_paq = ["SEQN", "PAD615"]
vars_smq = ["SEQN", "SMQ020"]
vars_depression = ["SEQN", "DPQ010", "DPQ020", "DPQ030", "DPQ040", "DPQ050", "DPQ060", "DPQ070", "DPQ080", "DPQ090"]

df_list = []
for y, c in zip(years, cycles):
    df_demo = load_nhanes_data(y, c, "DEMO", vars_demo)
    df_bmx = load_nhanes_data(y, c, "BMX", vars_bmx)
    df_thyroid = load_nhanes_data(y, c, "THYROD", vars_thyroid)

    # Fix for TSH column naming
    if "LBXTSH" not in df_thyroid.columns and "LBXTSH1" in df_thyroid.columns:
        df_thyroid.rename(columns={"LBXTSH1": "LBXTSH"}, inplace=True)

    df_slq = load_nhanes_data(y, c, "SLQ", vars_slq)
    df_paq = load_nhanes_data(y, c, "PAQ", vars_paq)
    df_smq = load_nhanes_data(y, c, "SMQ", vars_smq)
    df_dpq = load_nhanes_data(y, c, "DPQ", vars_depression)

    dfs = [df_demo, df_bmx, df_thyroid, df_slq, df_paq, df_smq, df_dpq]
    df_merged = dfs[0]
    for d in dfs[1:]:
        df_merged = pd.merge(df_merged, d, on="SEQN", how="outer")

    df_merged["cycle"] = c
    df_merged["year"] = y
    df_list.append(df_merged)

# Combine all years
df_final = pd.concat(df_list, ignore_index=True)

# Calculate PHQ-9 depression score (sum of 9 DPQ items)
phq_items = ["DPQ010", "DPQ020", "DPQ030", "DPQ040", "DPQ050", "DPQ060", "DPQ070", "DPQ080", "DPQ090"]
df_final["Depression"] = df_final[phq_items].sum(axis=1, skipna=False)

# Rename columns for clarity
df_final.rename(columns={
    "RIAGENDR": "Gender",
    "RIDAGEYR": "Age",
    "BMXBMI": "BMI",
    "LBXTSH": "TSH",
    "LBXTT3": "TotalT3",
    "LBXT4F": "FreeT4",
    "SLD010H": "SleepDuration",
    "SLQ060": "SleepQuality",
    "PAD615": "PhysicalActivity",
    "SMQ020": "SmokingStatus"
}, inplace=True)

# Create your final analysis dataframe with selected columns
df_analysis = df_final[[
    "SEQN",
    "Gender",
    "Age",
    "BMI",
    "TSH",
    "TotalT3",
    "FreeT4",
    "SleepDuration",
    "SleepQuality",
    "PhysicalActivity",
    "SmokingStatus",
    "Depression",
    "cycle",
    "year"
]].copy()

# Preview the dataframe
print(df_analysis.head())
```

## Data Cleaning and Handling of Missing Data

```python
# ------------------------
# NHANES Data Cleaning Script
# ------------------------

import numpy as np

# PHQ-9 items (Depression scale)
phq_items = ["DPQ010", "DPQ020", "DPQ030", "DPQ040", "DPQ050", "DPQ060", "DPQ070", "DPQ080", "DPQ090"]

# Categorical columns
cat_cols = ["Gender", "SmokingStatus", "SleepQuality"]
for col in cat_cols:
    df_final[col] = df_final[col].astype("category")

# Drop implausible SleepDuration and BMI values
df_final = df_final[(df_final["SleepDuration"] >= 0) & (df_final["SleepDuration"] <= 24)]
df_final = df_final[(df_final["BMI"] >= 10) & (df_final["BMI"] <= 80)]

# Recode invalid PHQ-9 values (9 ‚Üí NaN)
for item in phq_items:
    df_final[item] = df_final[item].replace(9, np.nan)

# Recode invalid PhysicalActivity values
invalid_pa_codes = [9999]
df_final["PhysicalActivity"] = df_final["PhysicalActivity"].replace(invalid_pa_codes, np.nan)

# Compute depression score (only if all items are present)
df_final["Depression"] = df_final[phq_items].sum(axis=1, skipna=False)

# Show missing data count before dropping
print("\nüìå Missing Data Before Cleaning:")
print(df_final[["SleepDuration", "BMI", "Depression", "TSH", "TotalT3", "FreeT4", "Age",
                "Gender", "SmokingStatus", "PhysicalActivity"]].isnull().sum().sort_values(ascending=False))

# Define key variables needed for modeling/statistics
key_vars = ["SleepDuration", "BMI", "Depression", "TSH", "TotalT3", "FreeT4",
            "Age", "Gender", "SmokingStatus", "PhysicalActivity"]

# Drop rows with missing data in any key variable
df_clean = df_final.dropna(subset=key_vars).reset_index(drop=True)

# Cap extreme hormone values (outliers)
df_clean.loc[df_clean["TSH"] > 20, "TSH"] = 20
df_clean.loc[df_clean["TotalT3"] > 300, "TotalT3"] = 300
df_clean.loc[df_clean["FreeT4"] > 4, "FreeT4"] = 4

# Show dataset shape and confirm all missing values removed
print(f"\n‚úÖ Final cleaned dataset shape: {df_clean.shape}")
print("\n‚úÖ Missing values after cleaning (should all be 0):")
print(df_clean[key_vars].isnull().sum())
```

##Shapiro-Wilk Normality Test

```python
from scipy.stats import shapiro
import pandas as pd

# Define numeric columns you want to test
numeric_cols = ["Age", "BMI", "TSH", "TotalT3", "FreeT4", "SleepDuration", "PhysicalActivity"]

# Create empty list to collect results
shapiro_results = []

# Run test for each variable
for col in numeric_cols:
    data = df_clean[col].dropna()
    if len(data) > 3:  # Shapiro requires at least 3 non-NaN values
        sample_data = data.sample(n=min(len(data), 5000), random_state=42)
        stat, p = shapiro(sample_data)
        interpretation = "Normal" if p > 0.05 else "Non-normal"
        shapiro_results.append({
            "Variable": col,
            "W Statistic": stat,
            "p-value": p,
            "Normality": interpretation
        })
    else:
        shapiro_results.append({
            "Variable": col,
            "W Statistic": None,
            "p-value": None,
            "Normality": "Insufficient data"
        })

# Create dataframe to display
norm_df = pd.DataFrame(shapiro_results)

# Format p-values for readability
norm_df["p-value"] = norm_df["p-value"].apply(lambda x: f"{x:.4f}" if pd.notnull(x) else "NA")

# Show the result
print("Shapiro-Wilk Normality Test Results:")
display(norm_df)


```

All variables failed the normality test (p < 0.05).

Most of them are strongly non-normal, especially TSH and FreeT4 (very low W values, consistent with extreme skewness)and Physical Activity (the most non-normal). Even Age and Sleep Duration, which looked mildly skewed, are statistically non-normal due to large sample size.

# T-Test

##Group Comparisons of Sleep Duration, Depression, and BMI by Gender and Smoking Status

```python
from scipy.stats import ttest_ind, f_oneway

# Variables
dependent_vars = ['SleepDuration', 'Depression', 'BMI']
categorical_IVs = {
    'Gender': [1, 2],
    'SmokingStatus': [1, 2]  # Assuming you‚Äôve already filtered out 7 and 9
}

def run_group_comparisons(df, dv_list, iv_dict):
    for dv in dv_list:
        for iv, valid_values in iv_dict.items():
            # Filter only valid groups (e.g., Gender: 1,2 or SmokingStatus: 1,2)
            df_filtered = df[df[iv].isin(valid_values)][[dv, iv]].dropna()
            groups = df_filtered[iv].unique()

            if len(groups) == 2:
                # Run t-test
                group1 = df_filtered[df_filtered[iv] == groups[0]][dv]
                group2 = df_filtered[df_filtered[iv] == groups[1]][dv]
                t_stat, p_val = ttest_ind(group1, group2, equal_var=False)
                print(f"t-test for {dv} by {iv}:")
                print(f"Groups: {groups[0]} vs {groups[1]}")
                print(f"t = {t_stat:.3f}, p = {p_val:.3f}\n")

            elif len(groups) > 2:
                # Run ANOVA
                group_data = [df_filtered[df_filtered[iv] == g][dv] for g in groups]
                f_stat, p_val = f_oneway(*group_data)
                print(f"ANOVA for {dv} by {iv}:")
                print(f"Groups: {groups}")
                print(f"F = {f_stat:.3f}, p = {p_val:.3f}\n")

            else:
                print(f"Skipped {dv} by {iv}: Not enough groups.\n")

# Run the tests
run_group_comparisons(df_clean, dependent_vars, categorical_IVs)
```

# One-Way ANOVA: Effect of Thyroid and Lifestyle Variables on Sleep, BMI, and Depression


```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Your DV and IV lists
dv_cols = ['SleepDuration', 'BMI', 'Depression']
iv_cols = ['TSH', 'TotalT3', 'FreeT4', 'Age', 'PhysicalActivity']  # Only continuous/multi-category IVs

print("\n===== ONE-WAY ANOVA RESULTS =====")
for dv in dv_cols:
    for iv in iv_cols:
        # Categorize continuous IV into 3 groups
        df_temp = df_clean.copy()
        df_temp[iv + '_cat'] = pd.qcut(df_temp[iv], q=3, labels=["Low", "Medium", "High"])

        # Fit ANOVA model
        model = ols(f"{dv} ~ C({iv + '_cat'})", data=df_temp).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)

        # Get p-value safely
        p_val = anova_table["PR(>F)"].iloc[0]
        print(f"{iv} (3 groups) ‚Üí {dv}: p = {p_val:.3f}")
```

Non-significant predictors (p ‚â• 0.05): TSH, TotalT3, FreeT4

Overall: Effects are small but meaningful, R¬≤ ‚âà 0.08

‚úÖ Overview:
R¬≤ values are low (1.6% for SleepDuration, 2.9% for BMI, 8% for Depression) ‚Üí predictors explain only a small portion of variance. Significant predictors generally have small effect sizes.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Dependent variables
dv_cols = ['SleepDuration', 'BMI', 'Depression']

# Independent variables
X = df_clean[['TSH', 'TotalT3', 'FreeT4', 'Age', 'PhysicalActivity']]
X = sm.add_constant(X)
# Dummy coding for categorical variables
X['Gender_2.0'] = (df_clean['Gender'] == 2).astype(int)
X['SmokingStatus_2.0'] = (df_clean['SmokingStatus'] == 2).astype(int)

# Set up figure
fig, axes = plt.subplots(len(dv_cols), 2, figsize=(16, 12))
fig.suptitle("Regression Model Diagnostics with R¬≤", fontsize=18, y=0.95)

for i, dv in enumerate(dv_cols):
    y = df_clean[dv]
    model = sm.OLS(y, X).fit()
    predicted = model.predict(X)
    residuals = y - predicted
    r_squared = model.rsquared

    # Predicted vs Actual
    sns.scatterplot(x=predicted, y=y, ax=axes[i,0])
    axes[i,0].plot([y.min(), y.max()], [y.min(), y.max()], color='red', linestyle='--')
    axes[i,0].set_xlabel("Predicted")
    axes[i,0].set_ylabel("Actual")
    axes[i,0].set_title(f"{dv}: Predicted vs Actual")
    axes[i,0].text(0.05, 0.95, f"R¬≤ = {r_squared:.3f}", transform=axes[i,0].transAxes,
                   fontsize=12, verticalalignment='top', bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.5))

    # Residual plot
    sns.scatterplot(x=predicted, y=residuals, ax=axes[i,1])
    axes[i,1].axhline(0, color='red', linestyle='--')
    axes[i,1].set_xlabel("Predicted")
    axes[i,1].set_ylabel("Residuals")
    axes[i,1].set_title(f"{dv}: Residuals vs Predicted")

plt.tight_layout()
plt.show()
```

# Section 10: Conclusions & Future Work

## Outcomes

Our analysis aimed to investigate the interplay between thyroid function, demographics, lifestyle, and key health outcomes using NHANES data from six survey cycles between 2007 to 2012. Our primary goal was to evaluate the association between thyroid hormones (TSH, Free T4, and Total T3) and three critical health indicators: sleep duration, Body Mass Index (BMI), and depression scores (PHQ-9). Furthermore, we sought to determine how important demographic and lifestyle factors (age, gender, physical activity, and smoking) influence or modify these core relationships. This investigation also carried out a crucial step, which was to describe the population-level distributions of these biomarkers and health outcomes to provide essential context for our models.  

We selected our biomarkers based on their established or potential links to thyroid function and lifestyle factors, which are known to impact physical and mental health outcomes. This helped us better understand the endocrine system‚Äôs contribution to public health.  

The analyses revealed that thyroid hormone levels (TSH, Total T3, Free T4) exhibited very weak or negligible associations with sleep duration, BMI, and depression in the general population. Correlation coefficients were near zero, indicating that typical variations in thyroid function have limited predictive value for these health outcomes. Thyroid hormone levels also showed skewness and non-normal distribution, highlighting the need for careful data preprocessing. Medication use related to sleep, obesity, or depression was low among participants, minimizing potential confounding effects.

**Sleep duration** was slightly influenced by TSH, gender, and physical activity. Higher TSH was associated with a small increase in sleep duration, while females and more physically active individuals slept slightly lesser. However, these effects were minimal, and overall variance explained was very low (R¬≤ ‚âà 0.016). Differences in sleep duration by gender or smoking status were not statistically significant.

**BMI** showed slight associations with Total T3, age, gender, and smoking status. Higher Total T3, older age, female gender, and non-smoking status were linked to slightly higher BMI. The predictors explained a small portion of variance (R¬≤ ‚âà 0.029), suggesting that while these factors are statistically significant, the effect sizes remain small.

**Depression scores** were influenced by age, physical activity, gender, and smoking status. Older participants and those with higher physical activity reported slightly lower depression scores, whereas females and smokers had higher scores. Predictors accounted for about 8% of variance in depression, indicating small but meaningful relationships.

Overall, demographic and lifestyle factors had more noticeable effects than thyroid hormones on BMI and depression, though their influence on sleep duration was weak. These results suggest that normal variations in thyroid function do not substantially affect sleep, BMI, or depression in the general population.

**However, we acknowledge the limitations of using NHANES:**

First, we had missing data that reduced our effective sample size, particularly for thyroid hormones Free T4 and Total T3 in the later cycles. This thus decreased the effectiveness of sample size and may have limited statistical power for some analyses.

Secondly, NHANES is a cross-sectional survey, which restricts our ability to infer causality or temporal relationships between thyroid hormone levels and health outcomes like sleep duration, BMI, or depression. Finally, certain variables, such as physical activity or sleep duration, rely on self-reported data, which may introduce reporting bias.

Despite these limitations, the NHANES dataset remains one of the most robust and nationally representative sources available, enabling findings to be generalized to the broader U.S. adult population.
